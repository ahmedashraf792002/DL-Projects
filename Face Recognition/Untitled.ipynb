```python
#pip install face-recognition
import face_recognition
import matplotlib.pyplot as plt
import os
import cv2
```

```python
def Rec(path):
    # Load known faces
    known_faces = []
    known_face_names = []
    known_faces_dir = "known_faces"
    for filename in os.listdir(known_faces_dir):
        image = face_recognition.load_image_file(os.path.join(known_faces_dir, filename))
        face_encoding = face_recognition.face_encodings(image)[0]
        known_faces.append(face_encoding)
        known_face_names.append(os.path.splitext(filename)[0])
    #unknown
    unknown_image = face_recognition.load_image_file(path)
    unknown_face_locations = face_recognition.face_locations(unknown_image)
    unknown_face_encodings = face_recognition.face_encodings(unknown_image, unknown_face_locations)
    for face_encoding, face_location in zip(unknown_face_encodings, unknown_face_locations):
        matches = face_recognition.compare_faces(known_faces, face_encoding)
        name = "Unknown"
        if True in matches:
            first_match_index = matches.index(True)
            name = known_face_names[first_match_index]
    return unknown_image,name
def add(image,name):
    path='known_faces/'+name+'.jpg'
    cv2.imwrite(path,image)
```

```python
unknown_image,name=Rec("test.jpeg")
plt.imshow(unknown_image)
plt.axis('off')
plt.title(f"Face Names: {name}")
plt.show()
```
